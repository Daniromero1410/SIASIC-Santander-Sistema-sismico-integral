```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 11: GRÃFICA 8 - COMPARATIVA SANTANDER VS OTROS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\nâš–ï¸ GENERANDO COMPARATIVA SANTANDER VS OTROS...")

# Agrupar por departamento
dept_counts = df.groupby('DEPARTAMENTO').agg({
    'MAGNITUD': ['count', 'mean', 'max'],
    'PROFUNDIDAD': 'mean'
}).round(2)
dept_counts.columns = ['Total_Sismos', 'Magnitud_Promedio', 'Magnitud_Max', 'Prof_Promedio']
dept_counts = dept_counts.sort_values('Total_Sismos', ascending=True).reset_index()

# Colores para cada departamento
colors = ['#6B46C1' if dept == 'Santander' else '#10B981' for dept in dept_counts['DEPARTAMENTO']]

fig = go.Figure()

fig.add_trace(go.Bar(
    y=dept_counts['DEPARTAMENTO'],
    x=dept_counts['Total_Sismos'],
    orientation='h',
    marker=dict(
        color=colors,
        line=dict(color='white', width=2)
    ),
    text=dept_counts['Total_Sismos'],
    textposition='outside',
    hovertemplate="<b>%{y}</b><br>Sismos: %{x}<br>Mag promedio: %{customdata[0]:.1f}<br>Prof promedio: %{customdata[1]:.1f} km<extra></extra>",
    customdata=dept_counts[['Magnitud_Promedio', 'Prof_Promedio']].values
))

fig.update_layout(
    title=dict(
        text='âš–ï¸ COMPARATIVA: SANTANDER VS OTROS DEPARTAMENTOS',
        font=dict(size=20, family='Arial Black', color='#6B46C1'),
        x=0.5,
        xanchor='center'
    ),
    xaxis_title='NÃºmero de Sismos',
    yaxis_title='Departamento/RegiÃ³n',
    height=500,
    plot_bgcolor='rgba(240,240,240,0.5)',
    paper_bgcolor='white',
    font=dict(family='Arial', size=14),
    showlegend=False
)

fig.show()
print("âœ… Comparativa generada")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 15: EXPORTACIÃ“N A SHAPEFILE (PARA ARCGIS PRO/ONLINE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\nğŸ—ºï¸ EXPORTANDO A SHAPEFILE...")

# Crear directorio de exportaciÃ³n
os.makedirs('exports_gis', exist_ok=True)

# Crear GeoDataFrame
geometry = [Point(xy) for xy in zip(df['LONGITUD'], df['LATITUD'])]
gdf = gpd.GeoDataFrame(df.copy(), geometry=geometry, crs="EPSG:4326")

# Preparar columnas para Shapefile (mÃ¡ximo 10 caracteres por nombre)
gdf_export = gdf.copy()
gdf_export = gdf_export.rename(columns={
    'FECHA_HORA': 'FECHA_HR',
    'PROFUNDIDAD': 'PROF_KM',
    'TIPO_PROFUNDIDAD': 'TIPO_PROF',
    'ES_SANTANDER': 'SANTANDER',
    'DEPARTAMENTO': 'DEPTO',
    'MES_NOMBRE': 'MES_NOM'
})

# Convertir datetime a string para Shapefile
gdf_export['FECHA_HR'] = gdf_export['FECHA_HR'].astype(str)
gdf_export['FECHA'] = gdf_export['FECHA'].astype(str)

# Exportar Shapefile
gdf_export.to_file('exports_gis/sismos_santander_2024.shp', driver='ESRI Shapefile', encoding='utf-8')

print("âœ… Shapefile exportado: exports_gis/sismos_santander_2024.shp")
print("   ğŸ“ Archivos generados: .shp, .shx, .dbf, .prj, .cpg")
print("   ğŸ¯ CRS: EPSG:4326 (WGS84)")
print("   ğŸ“Š Features: ", len(gdf_export))
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 16: EXPORTACIÃ“N A GEOJSON Y KML
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\nğŸŒ EXPORTANDO A GEOJSON Y KML...")

# Exportar GeoJSON
gdf_export.to_file('exports_gis/sismos_santander_2024.geojson', driver='GeoJSON')
print("âœ… GeoJSON exportado: exports_gis/sismos_santander_2024.geojson")

# Crear KML manualmente
kml_header = '''<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://www.opengis.net/kml/2.2">
<Document>
    <name>Sismos Santander 2024 - SIASIC</name>
    <description>Datos sÃ­smicos del departamento de Santander - 2024</description>

    <Style id="nido_sismico">
        <IconStyle><color>ff800080</color><scale>1.2</scale><Icon><href>http://maps.google.com/mapfiles/kml/paddle/purple-circle.png</href></Icon></IconStyle>
    </Style>
    <Style id="superficial">
        <IconStyle><color>ff0000ff</color><scale>1.0</scale><Icon><href>http://maps.google.com/mapfiles/kml/paddle/red-circle.png</href></Icon></IconStyle>
    </Style>
    <Style id="intermedio">
        <IconStyle><color>ff00a5ff</color><scale>1.0</scale><Icon><href>http://maps.google.com/mapfiles/kml/paddle/orange-circle.png</href></Icon></IconStyle>
    </Style>
    <Style id="profundo">
        <IconStyle><color>ffff0000</color><scale>1.0</scale><Icon><href>http://maps.google.com/mapfiles/kml/paddle/blu-circle.png</href></Icon></IconStyle>
    </Style>
'''

kml_placemarks = ""
for _, row in df.iterrows():
    style = row['TIPO_PROFUNDIDAD'].lower().replace(' ', '_').replace('Ã­', 'i')
    kml_placemarks += f'''
    <Placemark>
        <name>{row['MUNICIPIO']} - M{row['MAGNITUD']}</name>
        <description><![CDATA[
            <b>Fecha:</b> {row['FECHA']}<br>
            <b>Magnitud:</b> {row['MAGNITUD']}<br>
            <b>Profundidad:</b> {row['PROFUNDIDAD']:.1f} km<br>
            <b>Tipo:</b> {row['TIPO_PROFUNDIDAD']}<br>
            <b>RegiÃ³n:</b> {row['REGION']}
        ]]></description>
        <styleUrl>#{style}</styleUrl>
        <Point>
            <coordinates>{row['LONGITUD']},{row['LATITUD']},0</coordinates>
        </Point>
    </Placemark>'''

kml_footer = '''
</Document>
</kml>'''

with open('exports_gis/sismos_santander_2024.kml', 'w', encoding='utf-8') as f:
    f.write(kml_header + kml_placemarks + kml_footer)

print("âœ… KML exportado: exports_gis/sismos_santander_2024.kml")
print("   ğŸ“ Compatible con: Google Earth, ArcGIS, QGIS")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 19: INTRODUCCIÃ“N AL SIMULADOR SIASIC-AI
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("""
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ğŸ¤– SIASIC-AI: SIMULADOR INTELIGENTE DE SISMOS
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

Este mÃ³dulo implementa:

1ï¸âƒ£  MODELO DE OMORI-UTSU (PredicciÃ³n de RÃ©plicas)
    â€¢ Ley calibrada para el Nido SÃ­smico de Bucaramanga
    â€¢ PredicciÃ³n de frecuencia y probabilidad de rÃ©plicas
    â€¢ EstimaciÃ³n de magnitud mÃ¡xima de rÃ©plicas (Ley de BÃ¥th)

2ï¸âƒ£  MODELO DE ATENUACIÃ“N SÃSMICA
    â€¢ CÃ¡lculo de intensidad Mercalli por distancia
    â€¢ Zonas de afectaciÃ³n (daÃ±o severo, moderado, percepciÃ³n)
    â€¢ EstimaciÃ³n de poblaciÃ³n afectada

3ï¸âƒ£  SIMULADOR INTERACTIVO
    â€¢ SelecciÃ³n de epicentro, magnitud y profundidad
    â€¢ VisualizaciÃ³n de zonas de impacto en mapa
    â€¢ Reportes automÃ¡ticos de afectaciÃ³n

â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ğŸ“š TEORÃA:

LEY DE OMORI-UTSU:
    n(t) = K / (t + c)^p

    Donde:
    â€¢ n(t) = rÃ©plicas por dÃ­a en tiempo t
    â€¢ K = productividad (depende de magnitud)
    â€¢ c = constante de tiempo (0.1-0.2 dÃ­as para Nido)
    â€¢ p = tasa de decaimiento (~1.1 para Nido)

LEY DE BÃ…TH:
    M_max_rÃ©plica â‰ˆ M_principal - 1.2

â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
""")

# Importaciones adicionales necesarias
from scipy.stats import poisson

print("âœ… MÃ³dulo SIASIC-AI inicializado")
print("ğŸ“Š Ejecuta las siguientes celdas para configurar los modelos")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 20: MODELO DE OMORI-UTSU - PREDICCIÃ“N DE RÃ‰PLICAS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ”¬ CONFIGURANDO MODELO DE OMORI-UTSU...")
print("=" * 70)

class OmoriUtsuModel:
    """
    Modelo de Omori-Utsu para predicciÃ³n de rÃ©plicas sÃ­smicas.
    Calibrado para el Nido SÃ­smico de Bucaramanga.
    """

    def __init__(self):
        # ParÃ¡metros calibrados por tipo de profundidad
        self.params_by_depth = {
            'Superficial': {'c': 0.05, 'p': 1.0, 'k_factor': 1.2},
            'Intermedio': {'c': 0.08, 'p': 1.05, 'k_factor': 1.0},
            'Nido SÃ­smico': {'c': 0.15, 'p': 1.15, 'k_factor': 0.7},
            'Profundo': {'c': 0.20, 'p': 1.2, 'k_factor': 0.5}
        }

    def calculate_K(self, magnitude, depth_type='Nido SÃ­smico'):
        """Calcula productividad K basada en magnitud"""
        a, b = 1.0, -1.67
        k_factor = self.params_by_depth.get(depth_type, {}).get('k_factor', 1.0)
        return k_factor * (10 ** (a * magnitude + b))

    def get_params(self, depth_type):
        """Obtiene parÃ¡metros c y p segÃºn profundidad"""
        params = self.params_by_depth.get(depth_type, self.params_by_depth['Intermedio'])
        return params['c'], params['p']

    def aftershock_rate(self, t, magnitude, depth_type='Nido SÃ­smico'):
        """Tasa de rÃ©plicas en tiempo t (dÃ­as)"""
        K = self.calculate_K(magnitude, depth_type)
        c, p = self.get_params(depth_type)
        return K / ((t + c) ** p)

    def cumulative_aftershocks(self, t, magnitude, depth_type='Nido SÃ­smico'):
        """NÃºmero acumulado de rÃ©plicas hasta tiempo t"""
        K = self.calculate_K(magnitude, depth_type)
        c, p = self.get_params(depth_type)

        if p == 1:
            return K * np.log((t + c) / c)
        else:
            return (K / (1 - p)) * ((t + c)**(1-p) - c**(1-p))

    def predict_aftershocks(self, magnitude, depth_type='Nido SÃ­smico',
                           days=30, min_magnitude=2.0):
        """Predice rÃ©plicas para un perÃ­odo de tiempo"""
        results = []
        mag_factor = 10 ** (-(min_magnitude - 2.0))

        for day in range(1, days + 1):
            rate = self.aftershock_rate(day, magnitude, depth_type) * mag_factor
            cumulative = self.cumulative_aftershocks(day, magnitude, depth_type) * mag_factor
            prob = 1 - poisson.pmf(0, rate)

            results.append({
                'DIA': day,
                'TASA_REPLICAS': round(rate, 2),
                'ACUMULADO': round(cumulative, 2),
                'PROB_REPLICA_%': round(prob * 100, 1)
            })

        return pd.DataFrame(results)

    def max_aftershock_magnitude(self, mainshock_mag):
        """Magnitud mÃ¡xima de rÃ©plicas (Ley de BÃ¥th)"""
        return max(mainshock_mag - 1.2, 2.0)


# Instanciar modelo
omori_model = OmoriUtsuModel()

# Ejemplo de uso
print("\nğŸ“Š EJEMPLO: PredicciÃ³n para sismo M5.0 en el Nido SÃ­smico")
print("-" * 70)

pred_df = omori_model.predict_aftershocks(
    magnitude=5.0,
    depth_type='Nido SÃ­smico',
    days=14,
    min_magnitude=3.0
)

print(pred_df.to_string(index=False))

max_mag = omori_model.max_aftershock_magnitude(5.0)
print(f"\nğŸ¯ Magnitud mÃ¡xima esperada de rÃ©plicas: M{max_mag:.1f}")
print(f"ğŸ“ˆ Total rÃ©plicas esperadas (Mâ‰¥3.0) en 14 dÃ­as: {pred_df['ACUMULADO'].iloc[-1]}")
print("\nâœ… Modelo de Omori-Utsu configurado")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 22: MODELO DE ATENUACIÃ“N SÃSMICA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ¯ CONFIGURANDO MODELO DE ATENUACIÃ“N...")
print("=" * 70)

class SeismicAttenuationModel:
    """
    Modelo de atenuaciÃ³n sÃ­smica para calcular intensidad y zonas de afectaciÃ³n.
    Calibrado para la regiÃ³n de Colombia/Santander.
    """

    def __init__(self):
        # Coeficientes de atenuaciÃ³n (calibrados para Colombia)
        self.c1 = 2.5   # Constante
        self.c2 = 1.5   # Coeficiente de magnitud
        self.c3 = 1.8   # AtenuaciÃ³n geomÃ©trica
        self.c4 = 0.003 # AtenuaciÃ³n anelÃ¡stica

        # Escala de Mercalli
        self.mercalli_scale = {
            'I': {'min': 1, 'max': 1.9, 'desc': 'No sentido', 'color': '#FFFFFF'},
            'II': {'min': 2, 'max': 2.9, 'desc': 'Muy dÃ©bil', 'color': '#ACD8E9'},
            'III': {'min': 3, 'max': 3.9, 'desc': 'DÃ©bil', 'color': '#83D0DA'},
            'IV': {'min': 4, 'max': 4.9, 'desc': 'Ligero', 'color': '#7BC87D'},
            'V': {'min': 5, 'max': 5.9, 'desc': 'Moderado', 'color': '#F5F500'},
            'VI': {'min': 6, 'max': 6.9, 'desc': 'Fuerte', 'color': '#F5C800'},
            'VII': {'min': 7, 'max': 7.9, 'desc': 'Muy fuerte', 'color': '#FF9100'},
            'VIII': {'min': 8, 'max': 8.9, 'desc': 'Severo', 'color': '#E93C00'},
            'IX': {'min': 9, 'max': 9.9, 'desc': 'Violento', 'color': '#B40000'},
            'X': {'min': 10, 'max': 10.9, 'desc': 'Extremo', 'color': '#800000'},
            'XI': {'min': 11, 'max': 11.9, 'desc': 'CatastrÃ³fico', 'color': '#4B0082'},
            'XII': {'min': 12, 'max': 12, 'desc': 'ApocalÃ­ptico', 'color': '#000000'}
        }

    def calculate_intensity(self, magnitude, epicentral_dist, depth):
        """Calcula intensidad Mercalli en un punto"""
        R = np.sqrt(epicentral_dist**2 + depth**2)
        R = max(R, 1)
        intensity = self.c1 + self.c2 * magnitude - self.c3 * np.log10(R) - self.c4 * R
        return np.clip(intensity, 1, 12)

    def get_mercalli_level(self, intensity):
        """Convierte intensidad numÃ©rica a nivel Mercalli"""
        for level, info in self.mercalli_scale.items():
            if info['min'] <= intensity <= info['max']:
                return level, info
        return 'XII', self.mercalli_scale['XII']

    def calculate_affected_radius(self, magnitude, depth, target_intensity):
        """Calcula radio donde se alcanza una intensidad especÃ­fica"""
        r_min, r_max = 0.1, 1000

        while r_max - r_min > 0.1:
            r_mid = (r_min + r_max) / 2
            intensity = self.calculate_intensity(magnitude, r_mid, depth)

            if intensity > target_intensity:
                r_min = r_mid
            else:
                r_max = r_mid

        return r_mid

    def calculate_impact_zones(self, magnitude, depth):
        """Calcula zonas de impacto para diferentes intensidades"""
        return {
            'epicentro': {
                'radio_km': 0,
                'intensidad': self.calculate_intensity(magnitude, 0, depth),
                'color': '#800000',
                'descripcion': 'Epicentro'
            },
            'daÃ±o_severo': {
                'radio_km': self.calculate_affected_radius(magnitude, depth, 8),
                'intensidad': 'VIII+',
                'color': '#E93C00',
                'descripcion': 'DaÃ±o severo a estructuras'
            },
            'daÃ±o_moderado': {
                'radio_km': self.calculate_affected_radius(magnitude, depth, 6),
                'intensidad': 'VI-VII',
                'color': '#FF9100',
                'descripcion': 'DaÃ±o moderado, grietas'
            },
            'percepcion_fuerte': {
                'radio_km': self.calculate_affected_radius(magnitude, depth, 4),
                'intensidad': 'IV-V',
                'color': '#F5F500',
                'descripcion': 'PercepciÃ³n fuerte, objetos caen'
            },
            'percepcion_leve': {
                'radio_km': self.calculate_affected_radius(magnitude, depth, 2),
                'intensidad': 'II-III',
                'color': '#83D0DA',
                'descripcion': 'PercepciÃ³n leve'
            }
        }

    def haversine_distance(self, lat1, lon1, lat2, lon2):
        """Distancia entre dos puntos en km"""
        R = 6371
        lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
        dlat, dlon = lat2 - lat1, lon2 - lon1
        a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
        return R * 2 * np.arcsin(np.sqrt(a))


# Instanciar modelo
attenuation_model = SeismicAttenuationModel()

# Ejemplo
print("\nğŸ“Š EJEMPLO: Zonas de impacto para M5.0 a 147 km de profundidad")
print("-" * 70)

zones = attenuation_model.calculate_impact_zones(magnitude=5.0, depth=147)

for zone_name, zone_info in zones.items():
    print(f"\n{zone_name.upper().replace('_', ' ')}:")
    print(f"   Radio: {zone_info['radio_km']:.1f} km")
    print(f"   Intensidad: {zone_info['intensidad']}")
    print(f"   DescripciÃ³n: {zone_info['descripcion']}")

print("\nâœ… Modelo de atenuaciÃ³n configurado")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 24: FUNCIÃ“N PRINCIPAL DEL SIMULADOR
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ® CONFIGURANDO SIMULADOR...")

def clasificar_profundidad_sim(depth):
    """Clasifica el tipo de sismo por profundidad"""
    if depth < 70:
        return 'Superficial'
    elif depth < 140:
        return 'Intermedio'
    elif depth <= 180:
        return 'Nido SÃ­smico'
    else:
        return 'Profundo'

def simulate_earthquake(lat, lon, magnitude, depth):
    """
    FunciÃ³n principal del simulador.

    Args:
        lat: Latitud del epicentro
        lon: Longitud del epicentro
        magnitude: Magnitud del sismo
        depth: Profundidad en km

    Returns:
        dict con todos los resultados de la simulaciÃ³n
    """

    # Tipo de profundidad
    depth_type = clasificar_profundidad_sim(depth)

    # Zonas de impacto
    zones = attenuation_model.calculate_impact_zones(magnitude, depth)

    # Intensidad en epicentro
    epicenter_intensity = attenuation_model.calculate_intensity(magnitude, 0, depth)
    mercalli_level, mercalli_info = attenuation_model.get_mercalli_level(epicenter_intensity)

    # PoblaciÃ³n afectada por ciudad
    affected_cities = []
    for _, city in ciudades.iterrows():
        dist = attenuation_model.haversine_distance(lat, lon, city['LATITUD'], city['LONGITUD'])
        intensity = attenuation_model.calculate_intensity(magnitude, dist, depth)
        mercalli, _ = attenuation_model.get_mercalli_level(intensity)

        # Determinar zona
        zone = 'fuera_de_zona'
        for zname, zinfo in zones.items():
            if zname != 'epicentro' and dist <= zinfo['radio_km']:
                zone = zname
                break

        affected_cities.append({
            'CIUDAD': city['CIUDAD'],
            'POBLACION': city['POBLACION'],
            'DISTANCIA_KM': round(dist, 1),
            'INTENSIDAD': round(intensity, 1),
            'MERCALLI': mercalli,
            'ZONA': zone
        })

    affected_df = pd.DataFrame(affected_cities).sort_values('DISTANCIA_KM')

    # PredicciÃ³n de rÃ©plicas
    aftershocks = omori_model.predict_aftershocks(magnitude, depth_type, 14, 3.0)
    max_aftershock = omori_model.max_aftershock_magnitude(magnitude)

    # EnergÃ­a liberada
    energy_joules = 10**(1.5 * magnitude + 4.8)
    energy_tnt = energy_joules / 4.184e9

    return {
        'epicentro': {'lat': lat, 'lon': lon},
        'magnitud': magnitude,
        'profundidad': depth,
        'tipo_profundidad': depth_type,
        'intensidad_epicentro': round(epicenter_intensity, 1),
        'mercalli': mercalli_level,
        'mercalli_desc': mercalli_info['desc'],
        'zonas': zones,
        'ciudades_afectadas': affected_df,
        'replicas': aftershocks,
        'max_replica_mag': max_aftershock,
        'energia_joules': energy_joules,
        'energia_tnt': energy_tnt
    }

print("âœ… FunciÃ³n simulate_earthquake() lista")
print("ğŸ’¡ Uso: results = simulate_earthquake(lat, lon, magnitud, profundidad)")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA 30: COMPARATIVA DE ESCENARIOS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ“Š COMPARATIVA DE ESCENARIOS SÃSMICOS...")
print("=" * 70)

# Definir escenarios
escenarios = [
    {'nombre': 'Nido M4.5', 'lat': 6.78, 'lon': -73.18, 'mag': 4.5, 'depth': 147},
    {'nombre': 'Nido M5.0', 'lat': 6.78, 'lon': -73.18, 'mag': 5.0, 'depth': 147},
    {'nombre': 'Nido M5.5', 'lat': 6.78, 'lon': -73.18, 'mag': 5.5, 'depth': 147},
    {'nombre': 'Superficial M5.0', 'lat': 7.12, 'lon': -73.12, 'mag': 5.0, 'depth': 20},
    {'nombre': 'CrÃ­tico M6.5', 'lat': 6.80, 'lon': -73.15, 'mag': 6.5, 'depth': 145},
]

# Simular cada escenario
comparativa = []
for esc in escenarios:
    result = simulate_earthquake(esc['lat'], esc['lon'], esc['mag'], esc['depth'])

    pop_riesgo = result['ciudades_afectadas'][
        result['ciudades_afectadas']['ZONA'].isin(['daÃ±o_severo', 'daÃ±o_moderado'])
    ]['POBLACION'].sum()

    comparativa.append({
        'Escenario': esc['nombre'],
        'Magnitud': esc['mag'],
        'Prof_km': esc['depth'],
        'Tipo': result['tipo_profundidad'],
        'Mercalli': result['mercalli'],
        'Radio_DaÃ±o_km': result['zonas']['daÃ±o_severo']['radio_km'],
        'Pob_Riesgo': pop_riesgo,
        'Replicas_14d': result['replicas']['ACUMULADO'].iloc[-1]
    })

df_comp = pd.DataFrame(comparativa)
print("\nğŸ“‹ TABLA COMPARATIVA:")
print(df_comp.to_string(index=False))

# VisualizaciÃ³n
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('Radio de DaÃ±o Severo', 'PoblaciÃ³n en Riesgo', 'RÃ©plicas 14 dÃ­as')
)

colors = ['#22C55E', '#84CC16', '#EAB308', '#F97316', '#EF4444']

fig.add_trace(go.Bar(x=df_comp['Escenario'], y=df_comp['Radio_DaÃ±o_km'],
                      marker_color=colors, text=df_comp['Radio_DaÃ±o_km'].round(1),
                      textposition='outside'), row=1, col=1)

fig.add_trace(go.Bar(x=df_comp['Escenario'], y=df_comp['Pob_Riesgo'],
                      marker_color=colors,
                      text=[f'{v/1000:.0f}K' for v in df_comp['Pob_Riesgo']],
                      textposition='outside'), row=1, col=2)

fig.add_trace(go.Bar(x=df_comp['Escenario'], y=df_comp['Replicas_14d'],
                      marker_color=colors, text=df_comp['Replicas_14d'],
                      textposition='outside'), row=1, col=3)

fig.update_layout(
    title=dict(text='âš–ï¸ COMPARATIVA DE ESCENARIOS SÃSMICOS',
               font=dict(size=20, family='Arial Black', color='#6B46C1'), x=0.5),
    height=400, showlegend=False, paper_bgcolor='white'
)
fig.update_xaxes(tickangle=45)

fig.show()
```
